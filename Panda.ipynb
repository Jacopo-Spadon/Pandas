```python
# 1. show the movies with more than 7 in Rating & greater than 50 million Gross

df=movies_df_copy.dropna(subset=['Title'])

result_df_1 = df[['Title', 'Gross', 'Rating']]
result_rating = result_df_1[result_df_1['Rating'] >= 7]
result_df_1 = result_rating[result_rating['Gross'] >= 5e7]
print("List of movies with more than 7 in Rating & greater than 50 million Gross:") 
print(result_df_1)
```

```python
# 2. show the movies with more than 7 in Rating & greater than 50 million Gross & with Parental guidance as MPAA Rating

result_df_2 = df[['Title', 'MPAA Rating', 'Gross', 'Rating']]
result_rating = result_df_2[result_df_2['Rating'] >= 7]
result_df_1 = result_rating[result_rating['Gross'] >= 5e7]
result_df_2 = result_df_1[result_rating['MPAA Rating'] == "PG"]
print("List of movies with more than 7 in Rating & greater than 50 million Gross with Parental guidance as MPAA Rating:") 
print(result_df_2)
```

```python
# 4. show the list of top 5 movies based on Budget

top_5_budget=df.sort_values(by=['Budget'], ascending=False)
result_df_4 = top_5_budget[['Title', 'Budget']]
print("top 5 movies based on Budget:")
result_df_4.head(5)
```

```python
# 5. show the top 5 Comedy movies approved by the audience (use Rating)

result_df_5 = df.sort_values(by=['Rating'], ascending=False)
top_5_comedy = result_df_5[result_df_5['Genre'] == "Comedy"]
top_5_comedy = top_5_comedy[['Title', 'Rating']]
print("top 5 Comedy movies approved by the audience:")
top_5_comedy.head(5)
```

```python
# 6. top 5 movie names by Rating

result_df_5 = df.sort_values(by=['Rating'], ascending=False)
top_5_rating = result_df_5[['Title', 'Rating']]
print("top 5 movie names by Rating:")
top_5_rating.head(5)
```

```python
# 7. top 3 high Gross Romance movies produced after 2000 (for the date you can use a string)

result_df_7 = df.sort_values(by=['Release Date'], ascending=False)
top_romance = result_df_7[result_df_7['Genre'] == "Romance"]
top_3_romance = top_romance[top_romance['Release Date'].str.contains("20")]
top_3_romance = top_3_romance[['Title', 'Gross', 'Release Date']]
print("Top 3 high Gross Romance movies produced after 2000:")
top_3_romance.head(3)
```

```python
# 8. how many Genres are present in the dataframe? (use the function value_counts() which applies to Series, not Dataframe)
genres_df=pd.DataFrame(data=(df['Genre']))

s=genres_df.groupby(by=['Genre']).sum()
s
```

```python
# 9. top 5 expensive movies produced after 2000 (measured by Budget)

result_df_9 = df.sort_values(by=['Budget'], ascending=False)
top_5_budget = result_df_9[result_df_9['Release Date'].str.contains("20")]
top_5_budget = top_5_budget[['Title', 'Budget', 'Release Date']]
print("Top 5 expensive movies produced after 2000:")
top_5_budget.head(5)
```

```python
# 11. most and least expensive Genre (take an average of all Budget measures grouped by Genre - use groupBy() method)

result_df_11 = df.groupby(['Genre'])['Budget'].mean()

most_df_average_genre = result_df_11.sort_values(ascending=False).head(1)
least_df_average_genre = result_df_11.sort_values(ascending=False).tail(1)

print("The most expensive is", most_df_average_genre)

print("The least expensive is", least_df_average_genre)


```
