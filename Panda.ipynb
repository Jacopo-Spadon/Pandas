```python
# 1. show the movies with more than 7 in Rating & greater than 50 million Gross

df=movies_df_copy.drop([615])

result_df_1 = df[['Title', 'Gross', 'Rating']]
result_rating = result_df_1[result_df_1['Rating'] >= 7]
result_df_1 = result_rating[result_rating['Gross'] >= 5e7]
print("List of movies with more than 7 in Rating & greater than 50 million Gross:") 
print(result_df_1)
```

```python
# 2. show the movies with more than 7 in Rating & greater than 50 million Gross & with Parental guidance as MPAA Rating

result_df_2 = df[['Title', 'MPAA Rating', 'Gross', 'Rating']]
result_rating = result_df_2[result_df_2['Rating'] >= 7]
result_df_1 = result_rating[result_rating['Gross'] >= 5e7]
result_df_2 = result_df_1[result_rating['MPAA Rating'] == "PG"]
print("List of movies with more than 7 in Rating & greater than 50 million Gross with Parental guidance as MPAA Rating:") 
print(result_df_2)
```

```python
# 4. show the list of top 5 movies based on Budget

top_5_budget=df.sort_values(by=['Budget'], ascending=False)
result_df_4 = top_5_budget[['Title', 'Budget']]
print("top 5 movies based on Budget:")
result_df_4.head(5)
```

```python
# 5. show the top 5 Comedy movies approved by the audience (use Rating)

result_df_5 = df.sort_values(by=['Rating'], ascending=False)
top_5_comedy = result_df_5[result_df_5['Genre'] == "Comedy"]
top_5_comedy = top_5_comedy[['Title', 'Rating']]
print("top 5 Comedy movies approved by the audience:")
top_5_comedy.head(5)
```
